```python
# pip install pyecharts
# pycharts主题包 pip install echarts-themes-pypkg
# pyecharts全球国家地图 pip install echarts-countries-pypkg
import pyecharts as echarts
from wordcloud import WordCloud    # 词云包
from scipy.misc import imread
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import imageio as imageio
import re
import warnings
from decimal import Decimal
import tushare as ts

warnings.filterwarnings('ignore')
```

```python
# 从csv导入数据
df = pd.read_csv('data2018-11-06.csv')
# 图表的副标题
SUBTITLE = '数据截止2018.11.01'
```

```python
# 票房最高的电影Top10
# 按名字分组求和
df1 = df.sort_values(by='amount',ascending=False)
top10 = df1.head(10)
xData = top10['name'].tolist()
yData = top10['amount'].tolist()
# 画图
barTop10 = echarts.Bar('2018年电影票房Top10（万）',SUBTITLE)
barTop10.add('票房',xData,yData,xaxis_rotate=-45,is_legend_show=False,is_label_show=True)
grid = echarts.Grid()
grid.add(barTop10, grid_bottom="25%")
```

```python
# 人气最高的电影Top
# 按名字分组，对场均人次求平均值，并排序
df2 = df.sort_values(by='p_pc',ascending=False)
# 取top10
top10 = df2.head(10)
xData = top10['name'].tolist()
yData = top10['p_pc'].tolist()
# 画图
barTop10 = echarts.Bar('2018年电影人气Top10',SUBTITLE)
barTop10.add('场均人次',xData,yData,xaxis_rotate=-45,is_legend_show=False,is_label_show=True)
grid = echarts.Grid()
grid.add(barTop10, grid_bottom="18%")
```

```python
# 性价比最好的电影(根据价格，时长和口碑指数)
# 算法：口碑指数/最高口碑 * 50 + 最低价/价格 * 30 + (平均时长-abs(时长-平均时长))/平均时长 * 20
df3 = df[df['duration']>0] #删掉时长等于0的：因为豆瓣上有些电影没有时长，保存数据的时候以0保存的
# 按名字分组，对场均人次求平均值，并排序
# df2 = df.groupby(by='name')['p_pc'].mean().sort_values(ascending=False)
# 最高口碑
maxWom = df3['wom_index'].max()
# 最低价格
minPrice = df3['avg_price'].min()
# 平均时长
avgDuration = df3['duration'].mean()
# 计算单项得分
df3Score = pd.DataFrame()
for index, row in df3.iterrows():
    df3Score.set_value(index,'name',row['name'])
    df3Score.set_value(index,'wom_index',round(float(row['wom_index'] / maxWom * 50)))
    df3Score.set_value(index,'avg_price', round(float(minPrice / row['avg_price'] * 30)))
    df3Score.set_value(index,'duration',round(float(avgDuration - abs(row['duration']-avgDuration)) / avgDuration * 20))

df3Score['score_sum'] = round(df3Score.sum(axis=1),1)

# 计算性价比得分，并按得分排序
score = df3Score.sort_values(by=['score_sum'],ascending=False)
score
# 取top10
top10 = score.head(10)
attr = top10['name'].tolist()
tag = top10['score_sum'].tolist()
for i in range(len(attr)):
    attr[i] = attr[i] + ' '+str(tag[i])
v1 = top10['wom_index'].tolist()
v2 = top10['avg_price'].tolist()
v3 = top10['duration'].tolist()
barTop10 = echarts.Bar('2018年性价比最好的电影Top10',SUBTITLE)
barTop10.add("口碑得分", attr, v1,is_stack=True,is_label_show=True,label_pos='inside')
barTop10.add("价格得分", attr, v2,is_stack=True,is_label_show=True,label_pos='inside')
barTop10.add("时长得分", attr, v3,is_stack=True,is_label_show=True,label_pos='inside',xaxis_rotate=-45)
grid = echarts.Grid()
grid.add(barTop10, grid_bottom="26%")
```

```python
# 统计每个月有多少上映的电影数量，找出淡季和旺季
df4 = df
#将上映日期转换为日期类型
df4['list_date'] = pd.to_datetime(df4['list_date'])
# 获取每部电影的上映月份
for index, row in df4.iterrows():
    df4.set_value(index,'month',row['list_date'].month)
df4List = df4.groupby(by=['name','month']).size()
df4Month = df4List.to_frame('count')
resultList = df4Month.groupby(by='month').size()

xData = resultList.index.tolist()
for i in range(len(xData)):
    xData[i] = str(round(xData[i])) + '月'
yData = resultList.values.tolist()
# 画图
barMovieCount = echarts.Bar('2018年每月上映电影数量',SUBTITLE)
barMovieCount.add('上映电影数量',xData,yData,is_legend_show=False,is_label_show=True)
```

```python
# 评分 Top10
# 按评分倒序排列
data1 = df.sort_values(by='rate',ascending=False)
# 取前10条数据
top10 = data1.head(10)
xData = (top10['name']+'('+top10['region']+')').tolist() # 电影名称+制片地区作为x轴
yData = top10['rate'].tolist()
# 画图
barTop10 = echarts.Bar('2018年豆瓣评分Top10',SUBTITLE)
barTop10.add('豆瓣评分',xData,yData,xaxis_rotate=-45,is_legend_show=False,is_label_show=True)
grid = echarts.Grid()
grid.add(barTop10, grid_bottom="30%")
```

```python
# 按国家统计
# 按国家分组统计
data2 = df.groupby(by='region')
#取出分组后的数据
groupData = data2.size().reset_index(name='count')
data2Region = groupData['region']
mapValue = groupData['count']
# 国家/地区名称中英文转换
mapAttr = data2Region.map({'中国大陆':'China','香港':'China','台湾':'China','伊朗':'Iran','俄罗斯':'Russia',
                           '加拿大':'Canada','印度':'India','印度尼西亚':'Indonesia','德国':'Germany',
                           '意大利':'Italy','挪威':'Norway','日本':'Japan','法国':'France','波兰':'Poland',
                           '澳大利亚':'Australia','瑞士':'Switzerland','美国':'United States',
                           '英国':'United Kingdom','荷兰':'Netherlands','韩国':'Korea','马来西亚':'Malaysia'})
# 画图
map = echarts.Map("2018年各国家电影数量", SUBTITLE)
map.add(
    ' ',
    mapAttr,
    mapValue,
    maptype="world",
    is_visualmap=True,
    visual_text_color="#000",
    is_map_symbol_show=False,
    is_legend_show=False
)
```

```python
# 各类型电影占比
data3 = df['types'].tolist() # series转list
dataCount = len(df) # 数据量
typesStr = '|'.join(data3) # 将list所有字段拼接到一起
typesList = typesStr.split('|')
typesSeries = pd.value_counts(typesList) # 统计各种类型的数量
# 创建饼图
pie = echarts.Pie('各类型电影占比', SUBTITLE, title_pos='center')
style = echarts.Style()
pie_style = style.add(
    label_pos="center",
    is_label_show=True,
    label_text_color=None,
    is_legend_show=False
)
# 循环添加饼图(index:类型名，values:类型数量)
countPerLine = 7 # 每行显示5个
for i in range(len(typesSeries)):
    name = typesSeries.index[i] # 类型名称
    value = typesSeries.values[i] # 类型数量
    rate = round(value*100/dataCount,2) # 类型占比（按100份算，保留2位小数）
    pie.add(
        "", 
        [name, "其他"], 
        [rate, 100-rate], 
        center=[10+13*(i%countPerLine), 32+(i//countPerLine)*25], 
        radius=[18, 23], 
        **pie_style
    )
grid = echarts.Grid()
grid.add(pie)
```

```python
# 各国家/地区电影均分
data4 = df.groupby(by='region')
# 获取影名和数量
left = data4.size().reset_index(name='count')
# 获取均分
right = data4['rate'].mean().reset_index(name='avg')
# 找出left和right表的不同列，然后再进行合并，并按照均分倒序排列
cols_to_use = left.columns.difference(right.columns)
groupData = pd.merge(right, left[cols_to_use], left_index=True, 
                     right_index=True, how='outer').sort_values(by='avg',ascending=False).round(2)
data4Name = groupData['region'].tolist()
# data4Count = groupData['count'].tolist()
data4Avg = groupData['avg'].tolist()
# 画图
# scatter = echarts.Scatter("各国家/地区电影平均得分",SUBTITLE)
# scatter.add("平均分", data4Count, data4Avg)
barTop10 = echarts.Bar("各国家/地区电影平均得分",SUBTITLE)
barTop10.add('',data4Name,data4Avg,is_label_show=True,mark_line=["average"])
```

```python
# 各演员产量
data5 = df['actors'].dropna().tolist() # series转list
actorsStr = '|'.join(data5) # 将list所有字段拼接到一起
actorsList = actorsStr.split('|')
actorsSeries = pd.value_counts(actorsList) # 统计各个演员的电影数量
# 画图
# 自定义词云背景
bimg = imageio.imread('../resources/python.jpg')
wordcloud = WordCloud(background_color="white", mask=bimg, font_path='../resources/msyh.ttf', margin=10)
word_frequence = {index:value for index,value in actorsSeries.iteritems()}
wordcloud = wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file('imgCloud.jpg')
plt.close()
```
