# 猫眼（微信）、招行（掌上生活）、淘票票电影票价比较

```python
import pandas as pd
import numpy as np
import pymongo
```

```python
conn = pymongo.MongoClient('mongodb://localhost:27017/')
db = conn.moviePrice
```

```python
# 要剔除的特殊字符
str_replace_char = '[()（）：:"“”· ]'
```

## 猫眼电影数据

### 获取

```python
cursor_maoyan = db['maoyan_wechat'].find()
df_maoyan = pd.DataFrame(list(cursor_maoyan))
df_maoyan.sample(10)
```

```python
len(df_maoyan)
```

### 简单处理

```python
df_maoyan_format = pd.DataFrame(df_maoyan,columns=['cinema_name','movie_name','price','show_date','show_time'])
df_maoyan_format['cinema_name'] = df_maoyan_format['cinema_name'].str.replace(str_replace_char,'')
df_maoyan_format['movie_name'] = df_maoyan_format['movie_name'].str.replace(str_replace_char,'')
df_maoyan_format['show_time'] = df_maoyan_format['show_date']+' '+df_maoyan_format['show_time']
del df_maoyan_format['show_date']
df_maoyan_format['type']='maoyan'
df_maoyan_format.sample(10)
```

## 淘票票电影数据

### 获取

```python
cursor_taopiaopiao = db['taopiaopiao_doupiaowang'].find()
df_taopiaopiao = pd.DataFrame(list(cursor_taopiaopiao))
df_taopiaopiao.sample(10)
```

```python
len(df_taopiaopiao)
```

### 简单处理

```python
df_taopiaopiao_format = pd.DataFrame(df_taopiaopiao,columns=['cinema_name','movie_name','price','show_time'])
df_taopiaopiao_format['cinema_name'] = df_taopiaopiao_format['cinema_name'].str.replace(str_replace_char,'')
df_taopiaopiao_format['movie_name'] = df_taopiaopiao_format['movie_name'].str.replace(str_replace_char,'')
df_taopiaopiao_format['type']='taopiaopiao'
df_taopiaopiao_format.sample(10)
```

## 招行掌上生活电影数据

### 获取

```python
cursor_zhangshang = db['zhangshang'].find()
df_zhangshang = pd.DataFrame(list(cursor_zhangshang))
df_zhangshang.sample(10)
```

```python
len(df_zhangshang)
```

### 简单处理

```python
df_zhangshang_format = pd.DataFrame(df_zhangshang,columns=['cinema_name','movie_name','price','show_date','show_time'])
df_zhangshang_format['cinema_name'] = df_zhangshang_format['cinema_name'].str.replace(str_replace_char,'')
df_zhangshang_format['movie_name'] = df_zhangshang_format['movie_name'].str.replace(str_replace_char,'')
df_zhangshang_format['show_time'] = df_zhangshang_format['show_date']+' '+df_zhangshang_format['show_time']
del df_zhangshang_format['show_date']
df_zhangshang_format['type']='zhangshang'
df_zhangshang_format.sample(10)
```

## 数据拼合（取3个平台都有的数据）

```python
df_all = pd.concat([df_maoyan_format,df_taopiaopiao_format,df_zhangshang_format], ignore_index=True)
df_all['price'] = df_all['price'].astype('float64')
df_all['show_time'] = pd.to_datetime(df_all['show_time'], format='%Y-%m-%d %H:%M')
# 同一平台同一电影院同一电影同一时间的多个场次价格取最低
df_all_unique = df_all.iloc[df_all.sort_values(by='price',ascending=True).groupby(['cinema_name','movie_name','show_time','type']).apply(lambda x: x['price'].idxmin())]
# 获得3个平台都有的数据
df_result = df_all_unique.groupby(['cinema_name','movie_name','show_time']).filter(lambda x:len(x) == 3)
# pd.set_option('display.max_rows',100)
df_result.sample(10)
```

```python
print('数据数量：',len(df_taopiaopiao_format))
print('电影院数量：',len(df_taopiaopiao_format.groupby(by='cinema_name')))
print('电影数量：',len(df_taopiaopiao_format.groupby(by='movie_name')))
```

```python
print('数据数量：',len(df_zhangshang_format))
print('电影院数量：',len(df_zhangshang_format.groupby(by='cinema_name')))
print('电影数量：',len(df_zhangshang_format.groupby(by='movie_name')))
```
