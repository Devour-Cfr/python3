# 王者荣耀 5447条战绩数据分析

```python
import pymongo
import pyecharts
import pandas as pd
from wordcloud import WordCloud    # 词云包
from scipy.misc import imread
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import imageio as imageio

# 获取MongoDB连接
conn = pymongo.MongoClient('mongodb://localhost:27017/')
db = conn.pvp
coll = db['detailInfo']

coll.find_one()
```

## 时间范围

```python
cursor = coll.find({},{'eventtime':1,'_id':0},limit=1,sort=[("eventtime",pymongo.DESCENDING)])
cursor[0]
```

```python
cursor = coll.find({},{'eventtime':1,'_id':0},limit=1,sort=[("eventtime",pymongo.ASCENDING)])
cursor[0]
```

## 数据数量

```python
coll.estimated_document_count()
```

## 玩家数量

```python
cursor = coll.find({},{'acntcampBlue':1,'acntcampRed':1,'_id':0})
list_role = []
# 获取所有玩家的 roleID、roleName
for i in cursor:
    list_item = i['acntcampBlue'] + i['acntcampRed']
    for j in list_item:
        item = {}
        item['roleId']=j['roleId']
        item['roleName']=j['roleName']
        if item not in list_role:
            list_role.append(item)
len(list_role)
```

## 游戏模式

```python
cursor = coll.find({},{'mapName':1,'_id':0})
df_mapName = pd.DataFrame(list(cursor))
# 统计数量
sr_count = df_mapName.groupby(by='mapName').size()
# 排序
sr_count = sr_count.sort_values(ascending=False)
# 画图
attr = sr_count.index
value = sr_count.values
bar = pyecharts.Bar("各游戏模式场次数量")
bar.add("游戏模式", attr, value,xaxis_rotate=-45,is_legend_show=False,is_label_show=True)
grid = pyecharts.Grid()
grid.add(bar, grid_bottom="20%")
```

## 玩游戏时间统计

```python
cursor = coll.find({},{'eventtime':1,'_id':0})
# 转 dataframe
df_eventtime = pd.DataFrame(list(cursor)).sort_values(by=['eventtime'])
# 因为获取的数据中没有年份，所以给数据拼接个年份
df_eventtime['year'] = '2018-'
df_eventtime['eventtime'] = df_eventtime['year'].str.cat(df_eventtime['eventtime'])
del df_eventtime['year']
# 转为时间格式
df_eventtime['eventtime'] = pd.to_datetime(df_eventtime['eventtime'], format='%Y-%m-%d %H:%M')
# 将 eventtime 设置为索引
# df_eventtime.set_index('eventtime', inplace=True)
```

### 按小时统计

```python
df_eventtime['hour']=df_eventtime['eventtime'].dt.hour
df_hour = df_eventtime.groupby(by='hour').count()
# 画图
attr = df_hour.index
value = df_hour['eventtime']
line = pyecharts.Line("玩家在线时间统计（按小时）")
line.add("时间：H", attr, value, is_smooth=True,mark_point=["max", "min"])
```

### 按周统计

```python
df_eventtime['weekday']=df_eventtime['eventtime'].dt.weekday
df_week = df_eventtime.groupby(by='weekday').count()
# 画图
attr = ['周一','周二','周三','周四','周五','周六','周日']
value = df_week['eventtime']
line = pyecharts.Line("玩家在线时间统计（按周）")
line.add("时间：周", attr, value, is_smooth=True,mark_point=["max", "min"])
```

## 最热门英雄

```python
cursor = coll.find({},{'acntcampBlue':1,'acntcampRed':1,'_id':0})
list_hero = []
for i in cursor:
    list_item = i['acntcampBlue'] + i['acntcampRed']
    for j in list_item:
        item = {}
        item['heroName']=j['heroName']
        list_hero.append(item)
# list 转 series 并统计数量
sr_hero = pd.DataFrame(list_hero).groupby(by='heroName').size().sort_values(ascending=False)
```

```python
# 画条形图
# attr = sr_hero.index
# value = sr_hero.values
# bar = pyecharts.Bar("各游戏模式场次数量")
# bar.add("游戏模式", attr, value,xaxis_rotate=-45,is_legend_show=False,is_label_show=True)
# 自定义词云背景
bimg = imageio.imread('../resources/pvp.jpg')
wordcloud = WordCloud(background_color="white", mask=bimg, font_path='../resources/msyh.ttf', 
                      margin=10,max_font_size=130)
word_frequence = {index:value for index,value in sr_hero.iteritems()}
wordcloud = wordcloud.fit_words(word_frequence)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file('imgCloudPVP.jpg')
plt.close()
```
